```python
import matplotlib.pyplot as plt

x=[1,2,3,4]
y=[5,6,7,8]


x1=[7,5,3]
y1=[9,7,5]

plt.plot(x,y,label='line1')
plt.plot(x1,y1,label='line2')
plt.title('graphs')
plt.xlabel('x-axis')
plt.ylabel('y-axis')

plt.legend()
plt.show()
```

```python
plt.bar(x,y,color='r',label='bar 1')
plt.bar(x1,y1,color='g',label='bar 2')

plt.xlabel('x-axis')
plt.ylabel('y-axis')

plt.legend()
plt.show()
```
